```julia
using Plots, LaTeXStrings
```

```julia
x = range(0, stop=1, length=100)
x0 = sqrt(3)/3
rfun(x) = 2x/(1. + x^2)^2
sfun(x) = x^2 *(1. - x^2)/(1. + x^2)^2
a = 0.1; s = 0.01;
```

```julia
plot(rfun.(x), sfun.(x))
plot!([rfun(x0),], [sfun(x0),], seriestype=:scatter, legend=false)
plot!(r -> a*r)
xlabel!(L"$r$")
ylabel!(L"$s$")
```

```julia
s = 0.0;
b = 0.04;
plot(x -> b/a)
plot!(x -> s/x + x/(1 + x^2))
xlabel!(L"$x$")
ylabel!(L"$r$")
xlims!((0, 1))
ylims!((0, 1))
```
